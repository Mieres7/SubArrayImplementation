## SKA Mid and Low Array Filter
---

`io` function from `io2.py` allows you to build array with all antennas registered on a .cfg file

```python
from antenna_configs.io2 import io

ska_mid_file = "./antenna_configs/ska_mid_coordinates.cfg"
ska_low_file = "./antenna_configs/ska_low_coordinates.cfg"

mid_array = io(ska_mid_file)
low_array = io(ska_low_file)
```

SKA Low and Mid antennas can be selected by using the `get_ska_low_atennas_by_stage` and `get_ska_low_atennas_by_stage` methods, `stage` must be specified. Permitted stages are `AA4`,`AA*`, `AA2`, `AA1`, `AA0.5`. `AA*` is stablished as default.

```python
# SKA Mid AA* array
mid_array.get_ska_mid_atennas_by_stage("AA*")

# SKA Low AA* array
low_array.get_ska_low_atennas_by_stage("AA*")
```

Use the `plot_array_layout` function to visualize the obtained array

```python
mid_array.plot_array_layout()
```

```python
low_array.plot_array_layout()
```

### Filter Classes
An aditional class can be used to filter the antennas in the Array. `Filter` class is a abstract class that allos others to implement the `filter` method.

```python
from Filter.name_filter import IdFilter
```

For example `IdFilter Class`, allows to filter in two modes.
- `exact` [default]: returns an Array with exact antennas specified
- `partial`: returns an Array with antennas that contains the string specified

```python
id_filter_exact_mode = IdFilter(ids=["SKA001","SKA063","M002","M021"])
id_filter_partial_mode = IdFilter(ids=["M"], filter_mode="partial")

personal_array = id_filter_exact_mode.filter(array=mid_array)
aastar_with_only_merkat = id_filter_partial_mode.filter(array=mid_array)
```

```python
aastar_with_only_merkat.plot_array_layout(title="SKA Mid AA* - MERKAT")
```

```python
personal_array.plot_array_layout(title="SKA Mid AA* - Personalized")    
```

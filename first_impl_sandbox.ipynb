## SKA Mid and Low Array Filter
---

MidSubArray and LowSubArray classes allows you to build array depending on SKA stage. Permitted stages are `AA4`,`AA*`, `AA2`, `AA1`, `AA0.5`. `AA*` is stablished as default.

```python
from Array.mid_subarray import MidSubArray
from Array.low_subarray import LowSubarray
```

Default usage of Mid and Low classes

```python
mid_aastar_array = MidSubArray()
low_aastar_array = LowSubarray()
```

Mid and Low classes specifing SKA stage

```python
mid_aa2_array = MidSubArray("AA2")
low_aa05_array = LowSubarray("AA0.5")
```

Use the `plot_array_layout` function to visualize the obtained array

```python
mid_aastar_array.plot_array_layout(scale="kilo", title="SKA Mid AA* Layout")
```

```python

low_aastar_array.plot_array_layout(scale="kilo", title="SKA Low AA* Layout")
print(low_aastar_array.antenna_data)
```

### Filter Classes
An aditional class can be used to filter the antennas in the Array. `Filter` class is a abstract class that allos others to implement the `filter` method.

```python
from Filter.name_filter import IdFilter
```

For example `IdFilter Class`, allows to filter in two modes.
- `exact` [default]: returns an Array with exact antennas specified
- `partial`: returns an Array with antennas that contains the string specified

```python
id_filter_exact_mode = IdFilter(ids=["SKA001","SKA063","M002","M021"])
id_filter_partial_mode = IdFilter(ids=["M"], filter_mode="partial")

personal_array = id_filter_exact_mode.filter(array=mid_aastar_array)
aastar_with_only_merkat = id_filter_partial_mode.filter(array=mid_aastar_array)
```

```python
aastar_with_only_merkat.plot_array_layout(title="SKA Mid AA* - MERKAT")
```

```python
personal_array.plot_array_layout(title="SKA Mid AA* - Personalized")    
```
